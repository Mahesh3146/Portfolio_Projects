## **Bivariate Distribution**

```python
sns.jointplot(auto["engine_size"],auto["horsepower"])
 
```

```python
sns.jointplot(auto["horsepower"],auto["price"])
```

## **Hex Bin Plot**

```python
sns.jointplot(auto["horsepower"],auto["price"],kind = "hex")
```

```python
sns.jointplot(auto["peak_rpm"],auto["price"],kind = "hex")
```

# **Reg Bin Plots**

```python
sns.jointplot(auto["horsepower"],auto["price"],kind = "reg")
```

```python
sns.jointplot(auto["engine_size"],auto["price"],kind = "reg")
```

# **Resid Bin Plots**

```python
sns.jointplot(auto["horsepower"],auto["price"],kind = "resid")
```

```python
sns.jointplot(auto["stroke"],auto["horsepower"],kind = "resid")
```

# **Kernal Density Estimator Bin Plot**

```python
sns.jointplot(auto["horsepower"],auto["price"],kind = "kde")
```

```python
sns.jointplot(auto['engine_size'],auto['horsepower'],kind = "kde")
```
